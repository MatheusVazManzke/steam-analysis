```python
steam_df.dropna(subset=['Screenshots'], inplace=True)
steam_df.dropna(subset=['Movies'], inplace=True)
steam_df['Tags'].fillna('empty', inplace=True)
steam_df['About the game'].fillna('empty', inplace=True)

steam_df.dropna(subset=['Genres'], inplace=True)
```

```python
steam_df['about_length'] = steam_df['About the game'].apply(lambda x: len(x))
```

```python
steam_df['Supported languages'] = steam_df['Supported languages'].apply(lambda x: x.replace("[", "").replace("]", ""))
steam_df['Supported languages'] = steam_df['Supported languages'].apply(lambda x: set(x.split(',')))
steam_df['n_screens'] = steam_df['Screenshots'].apply(lambda x: len(set(x.split(','))))
steam_df['n_movies'] = steam_df['Movies'].apply(lambda x: len(set(x.split(','))))
steam_df['n_tags'] = steam_df['Tags'].apply(lambda x: len(set(x.split(','))))
steam_df['n_languages'] = steam_df['Supported languages'].apply(lambda n: len(n))


```

```python
steam_df['has_publisher'] = steam_df['Developers'] != steam_df['Publishers']
steam_df['perceived_quality'] = steam_df['Positive'] / (steam_df ['Negative'] + 1)
steam_df['no_user_reactions'] = ((steam_df['Positive'] + steam_df ['Negative']) == 0)
steam_df['total_reviews'] = (steam_df['Positive'] + steam_df ['Negative'])
```

```python
steam_df['has_support_email'] = ~steam_df['Support email'].isna()
steam_df['has_support_url'] = ~steam_df['Support url'].isna()
steam_df['has_website'] = ~steam_df['Website'].isna()
```

```python
# Convert the 'date_string' column to datetime format
steam_df['Release date'] = pd.to_datetime(steam_df['Release date'], format='mixed')

# Extract month, day, and year into separate datetime columns
steam_df['month'] = steam_df['Release date'].dt.month
steam_df['day'] = steam_df['Release date'].dt.day
steam_df['year'] = steam_df['Release date'].dt.year
```

```python
steam_df.drop(columns='AppID', inplace=True)
```

```python
steam_df['success_1000_threshold'] = (steam_df['total_reviews']>1000).astype(int)
steam_df['success_500_threshold'] = (steam_df['total_reviews']>500).astype(int)
```
